```python
import numpy as np

Resources_Matrix = np.array([[100, 200], [150, 250], [200, 300]])
Allocation_Factors = np.array([[1.1, 0.9], [1.2, 1.0]])

Result = np.dot(Resources_Matrix, Allocation_Factors)

print("Resulting Matrix:\n", Result)
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

forecast_values = np.array([0.5, 1.0, -1.0, 2.0, -2.0])

sigmoid_output = sigmoid(forecast_values)

print("Sigmoid Output:\n", sigmoid_output)
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_gradient(x):
    sigmoid_value = sigmoid(x)
    return sigmoid_value * (1 - sigmoid_value)

input_values = np.array([0.5, 1.0, -1.0, 2.0, -2.0])

sigmoid_gradient_output = sigmoid_gradient(input_values)

print("Sigmoid Gradient Output:\n", sigmoid_gradient_output)
```
